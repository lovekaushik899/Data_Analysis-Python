```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("thermodynamics.csv")
```

```python
data
```

```python
x1=data.drop(columns=['Volume'])
y1=data.Volume
```

```python
x1
```

```python
y1
```

```python
x1_train, x1_test, y1_train, y1_test = train_test_split(x1, y1, test_size=0.2, random_state=42)
```

```python
x1_train
```

```python
x1_test
```

```python
y1_train
```

```python
y1_test
```

```python
regressor = DecisionTreeRegressor(random_state=42)
```

```python
regressor
```

```python
regressor.fit(x1_train, y1_train)
```

```python
y1_pred = regressor.predict(x1_test)
```

```python
y1_pred
```

```python
mse = mean_squared_error(y1_test, y1_pred)
```

```python
mse
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(y1_test, y1_pred, color='blue', label='Predicted')
plt.plot([min(y1_test), max(y1_test)], [min(y1_test), max(y1_test)], 'r--', lw=2, label='Prediction')
plt.title('Decision Tree: Actual vs Predicted Volume')
plt.xlabel('Actual Volume')
plt.ylabel('Predicted Volume')
plt.legend()
plt.show()
```
