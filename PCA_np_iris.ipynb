```python
#Dividing the data into X and Y variables
x=data[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']]
y=data['Species']
```

```python
#For generating the heatmap of the data
#x.corr() means finding the correlation between the columns od X
#annot=True means enabling the annotation of the output
sns.heatmap(x.corr(), annot=True)
```

```python
#Applying the StandardScaler for scaling the dataset
scaler=StandardScaler()
```

```python
scaler
```

```python
#For fitting X accorfing to the scaler and scaling it
x_scaled=scaler.fit_transform(x)
```

```python
#Print the scaled data of X
x_scaled
```

```python
#Transposing the features on the scaled data of X
features=x_scaled.T
```

```python
features
```

```python
covariance_matrix=np.cov(features)
```

```python
covariance_matrix
```

```python
#linalg stands for Linear Algebra
#eig stands for eigen
values, vectors=np.linalg.eig(covariance_matrix)
```

```python
#Print the eigen vales
#Contribution of eigen values =(Eigen Value/Sum of Eigen Values)
values
```

```python
#Print the eigen vectors 
vectors
```

```python
#Print the shape of the eigen values
values.shape
```

```python
#Print the shape of the eigen vectors
vectors.shape
```

```python
#For calculating the percentage contribution of the eigen values
percentage_variance=[]
for i in range(len(values)):
    percentage_variance.append(values[i]/np.sum(values))
```

```python
#Print the percentage contribution of the eigen values
percentage_variance
```

```python
#Since 1st and 2nd vectors are maxing most of the contributions, we will take only the first 2 rows of the eigen vectors
#PC1 and PC2 will have 1st and 2nd vectors
proj_1=x_scaled.dot(vectors.T[0])
proj_2=x_scaled.dot(vectors.T[1])
result=pd.DataFrame(proj_1, columns=['PC1'])
result['PC2']=proj_2
```

```python
result
```

```python
sns.heatmap(result.corr(), annot=True)
```

```python
#Since 1st and 2nd vectors are maxing most of the contributions, we will take only the first 2 rows of the eigen vectors
#PC1 and PC2 will have 1st and 2nd vectors
proj_1=x_scaled.dot(vectors.T[0])
proj_2=x_scaled.dot(vectors.T[1])
result=pd.DataFrame(proj_1, columns=['PC1'])
result['PC2']=proj_2
result['species']=y
```

```python
result
```

```python
#Generate the scatter plot for the PCA with respect to the Species
plt.figure(figsize=(10, 6))
sns.scatterplot(x=result['PC1'], y=result['PC2'], hue='species', data=result, palette='Set1', s=100)
plt.title('Principal Component Analysis of Iris Dataset')
plt.xlabel('1st Principal Component')
plt.ylabel('2nd Principal Component')
plt.legend(title='Species')
```

```python
slope, intercept=np.polyfit(result['PC1'], result['PC2'], 1)
x_vals=np.linspace(min(result['PC1']), max(result['PC2']), 100)
y_vals=slope*x_vals+intercept
```
